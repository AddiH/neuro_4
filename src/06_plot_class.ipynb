```python
import pickle
import numpy as np
import matplotlib.pyplot as plt

f = open('../data/PvsN_05_permutation.pkl', 'rb')
score_cv_test, scores_perm, pvalue = pickle.load(f)
f.close()

f = open('../data/PvsN_02_searchlight_prep.pkl', 'rb')
fmri_img_train, fmri_img_test, idx_train,idx_test, conditions_train, conditions_test = pickle.load(f)
f.close()

# How many classes
n_classes = np.unique(conditions_test).size
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Assuming 'conditions_test', 'scores_perm', 'score_cv_test', and 'pvalue' are predefined

n_classes = np.unique(conditions_test).size

# Set the style of the plot to something more contemporary
plt.style.use('ggplot')

# Adjust the figure size to give our plot more room
plt.figure(figsize=(7, 6))

# Plot the histogram
plt.hist(scores_perm, bins=20, alpha=0.7, label='Permutation Scores', color='skyblue', linewidth=2)

# Set plot limits
ylim = plt.ylim()

# Plot for classification score with a line
plt.axvline(x=score_cv_test, color='green', linestyle='--', linewidth=2, label=f'Classification Score (p-value: {pvalue:.3f})')

# Plot for chance level with a line
plt.axvline(x=1. / n_classes, color='orange', linestyle='--', linewidth=2, label='Chance Level')

# Set the new limits after plotting
plt.ylim(ylim)

# Enhance the readability and layout
plt.title("Permutation Test Score Distribution \n Positive vs Negative", fontsize=16)
plt.xlabel('Score', fontsize=14)
plt.ylabel('Frequency', fontsize=14)

# Place a legend below the plot
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), fancybox=True, shadow=True, ncol=1)

# Adjust layout to accommodate all elements, including the external legend
plt.tight_layout(rect=[0, 0.1, 1, 1])

# Save and show the plot
plt.savefig("../final_plots/permutation_pos_neg.png", dpi=300)
plt.show()
```
